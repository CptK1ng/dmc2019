# Creating new feature totalScannedLineItems

```python
df_train = pd.read_csv("../data/train.csv", sep="|")
df_test = pd.read_csv("../data/test.csv", sep="|")
df_train.head()
```

```python
df_train['totalScannedLineItems'] =  df_train['totalScanTimeInSeconds'] * df_train['scannedLineItemsPerSecond']
df_test['totalScannedLineItems'] =  df_test['totalScanTimeInSeconds'] * df_test['scannedLineItemsPerSecond']
df_train.head()
```
